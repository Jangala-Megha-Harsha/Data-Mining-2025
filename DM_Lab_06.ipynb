**Use PCA (Principle Component Analysis) to reduce dimensions of a data. <br>
Visualize the reduced features**

```python
# Importing Required Libraries
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
# Step-1: Load the Iris Dataset
iris = load_iris()
X = iris.data  # Features
y = iris.target  # Labels
```

```python
X
```

```python
y
```

```python
X.shape
```

```python
y.shape
```

```python
# Step-2: Preprocess the Data (Standardization)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
X_scaled.shape
```

```python
# Step-3: Apply PCA to reduce to 2 dimentions
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca
```

```python
X_pca.shape
```

```python
# Step-4: Visualize the reduced features (2D scatter plot)
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=100)
plt.title("PCA of Iris Dataset (2 Components)")
plt.xlabel("Principle Component 1")
plt.ylabel("Principle Component 2")
plt.colorbar(label='Target Class')
plt.show()
```

```python
# Step-4: Visualize the reduced features (2D scatter plot)
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=150)
plt.title("PCA of Iris Dataset (2 Components)")
plt.xlabel("Principle Component 1")
plt.ylabel("Principle Component 2")
plt.colorbar(label='Target Class')
plt.show()
```

```python
# Step-4: Visualize the reduced features (2D scatter plot)
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=50)
plt.title("PCA of Iris Dataset (2 Components)")
plt.xlabel("Principle Component 1")
plt.ylabel("Principle Component 2")
plt.colorbar(label='Target Class')
plt.show()
```

```python
# Optional: Explained Varience Ratio
print("Explained Varience Ratio of Each Component: ")
print(pca.explained_variance_ratio_)
```
