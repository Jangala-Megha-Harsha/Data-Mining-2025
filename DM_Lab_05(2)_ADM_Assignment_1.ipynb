**Performancing Feature Scaling**

**1.**

**(a) Using MinMaxScaler (Built-in Library) Min-Max Scaling transforms data to a range of [0, 1]**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df[['Age', 'Salary']] = scaler.fit_transform(df[['Age', 'Salary']])
```

**(b) Manually Implementing Min-Max Scaling Min-Max Scaling formula:**

```python
df['Job_Satisfaction'] = (df['Job_Satisfaction'] - df['Job_Satisfaction'].min()) / (df['Job_Satisfaction'].max() - df['Job_Satisfaction'].min())
```

**2. (a) Standard Scaling (Z-Score Normalization) Standard Scaling transforms features to have a mean of 0 and a standard deviation of 1.**

```python
from sklearn.preprocessing import StandardScaler

s_scaler = StandardScaler()
```

```python
df[['Work_Hours_Per_Week']] = s_scaler.fit_transform(df[['Work_Hours_Per_Week']])
```

```python
df["Work_Hours_Per_Week"]
```

```python
df['Salary'] = (df['Salary'] - df['Salary'].mean()) / df['Salary'].std()
```
