## **1. Employee Salary Analysis**
A company wants to analyze its employeesâ€™ salaries to determine increments based on performance. The HR department has collected the following dataset:


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "Employee": ["John", "Sarah", "Mike", "Emma", "Tom"],
    "Department": ["It", "HR", "IT", "Finance", "HR"],
    "Salary": [60000, 50000, 75000, 65000, 48000],
    "Experience (Years)": [5, 3, 7, 4, 2],
    "Performance Rating (Out of 5)": [4.5, 3.8, 4.9, 4.2, 3.5]
}
```

Q1 : Create a DataFrame using the given Data

```python
df = pd.DataFrame(data)
df
```

Q2 : Calculate the average salary of employees in each department.

```python
avg_salary_by_department = df.groupby('Department')['Salary'].mean()
avg_salary_by_department
```

Q3 : Filter out employees who have a performance rating above 4.0.

```python
top_performaers = df[df['Performance Rating (Out of 5)'] > 4.0]
top_performaers
```

Q4 : Add a new column called "Increment", where employees with a rating above 4 get a 10% increment, and others get a 5% increment.

```python
df['Increment'] = np.where(df['Performance Rating (Out of 5)'] > 4, 0.10, 0.05)
df
```

```python
df.columns
```

Q5 : Sort employees in descending order of salary after the increment.

```python
df['New Salary'] = df['Salary'] + df['Salary'] * df['Increment']
df
```

```python
df_sorted = df.sort_values('New Salary', ascending=False)
df_sorted
```

## **2. COVID-19 Cases Analysis**
A health department is monitoring COVID-19 cases in different cities. They have recorded the following data:


```python
data_2 = {
    "city": ["New York", "Los Angeles", "Chicago", "Houston", "Miami"],
    "total_cases": [100000, 85000, 65000, 70000, 55000],
    "recovered": [75000, 70000, 50000, 60000, 45000],
    "deaths": [5000, 3500, 2500, 2000, 1500]
}
```

Q1 : Create a DataFrame for the given data.

```python
df2 = pd.DataFrame(data_2)
df2
```

Q2 : Calculate the recovery rate ((Recovered / Total Cases) * 100) for each city and add it as a new column.

```python
df2['recovery_rate'] = (df2['recovered'] / df2['total_cases']) * 100
df2
```

Q3 : Identify cities where the death rate ((Deaths / Total Cases) * 100) is higher than 5%.

```python
df2['death_rate'] = (df2['deaths'] / df2['total_cases']) * 100

high_death_rate_cities = df2[df2['death_rate'] >= 5.0]['city']
high_death_rate_cities
```

Q4 : Sort the cities by the highest recovery rate.

```python
sorted_df2 = df2.sort_values('recovery_rate', ascending=False)
sorted_df2
```

Q5 : Find the total number of cases and deaths across all cities.

```python
total_cases = df2['total_cases'].sum()
total_deaths = df2['deaths'].sum()

print("Total Cases:", total_cases)
print("Total Deaths:", total_deaths)
```
